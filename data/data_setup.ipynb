```python
import scanpy as sc
import anndata as ad
import numpy as np
import pandas as pd
```

```python
#q = sc.read_h5ad('CRUK_storming_cancer_gastric_all_cells_2022_v2.h5ad')
```

```python
data_prefix = '/users/dgibbs/Work/CRUK/'
```

```python
barcodes_f = pd.read_csv(data_prefix+"gastric_fibroblasts/barcodes.tsv.gz", sep="\t", header=None)
features_f = pd.read_csv(data_prefix+"gastric_fibroblasts/features.tsv.gz", sep="\t", header=None, index_col=1)
matrix_f = sc.read_mtx(data_prefix+"gastric_fibroblasts/matrix.mtx.gz")
meta_f = pd.read_csv(data_prefix+"gastric_fibroblasts/meta.tsv", sep="\t")
```

```python
features_f.columns = ['gene_id']
barcodes_f.columns = ['Cell_ID']

meta_f.index.name = None
features_f.index.name = None
barcodes_f.index.name = None
```

```python
meta_f
```

```python
qfib = sc.AnnData(X=matrix_f.X.transpose(), obs=meta_f, var=features_f)
```

```python
qfib
```

```python
barcodes_e = pd.read_csv(data_prefix+"gastric_epithelial/barcodes.tsv.gz", sep="\t", header=None)
features_e = pd.read_csv(data_prefix+"gastric_epithelial/features.tsv.gz", sep="\t", header=None, index_col=1)
matrix_e = sc.read_mtx(data_prefix+"gastric_epithelial/matrix.mtx.gz")
meta_e = pd.read_csv(data_prefix+"gastric_epithelial/meta.tsv", sep="\t")
```

```python
features_e.columns = ['gene_id']
barcodes_e.columns = ['Cell_ID']

meta_e.index.name = None
features_e.index.name = None
barcodes_e.index.name = None
```

```python
meta_e
```

```python
qepi = sc.AnnData(X=matrix_e.X.transpose(), obs=meta_e, var=features_e)
```

```python
#sc.pp.subsample(data=qepi, fraction=0.4)
```

```python
qepi
```

```python
q = ad.concat( [qfib, qepi] , join='outer')
```

```python
q.obs_names_make_unique()
q.var.index = q.var.index.astype(str)
q.obs['Cell_ID'] = q.obs['Cell_ID'].astype(str)
```

```python
q.obs
```

```python
q.write_h5ad(data_prefix + 'gastric_fibs_and_epi0_4.h5ad')
```
